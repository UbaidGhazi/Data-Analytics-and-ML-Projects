```python
import seaborn as sns
import matplotlib .pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv("C:\\Users\\HP\\Downloads\\Zomato data  (1).csv")
print(dataframe)
```

```python
dataframe
```

```python
def handleRate(value):
    value=str(value).split("/")
    value=value[0];
    return float(value)

dataframe["rate"]=dataframe["rate"].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info
```

```python
sns.countplot(x=dataframe["listed_in(type)"])
plt.xlabel("Types of Resturant")
plt.show()
```

```python
dataframe.head()
```

```python
grouped_data=dataframe.groupby("listed_in(type)")["rate"].sum()
result=pd.DataFrame({"vote":grouped_data})
plt.plot(result,c="g",marker="o")
plt.xlabel("Type of Resturant",c="m",size="20")
plt.ylabel("vote",c="y",size="20")
plt.show()
```

```python
plt.hist(dataframe["rate"],bins=4,color="y")
plt.show()
```

```python
Couple_data=dataframe["approx_cost(for two people)"]
sns.countplot(x=Couple_data)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x="online_order",y="rate",data=dataframe,hue="online_order")
plt.show()
```

```python
pivot_table=dataframe.pivot_table(index="listed_in(type)",columns="online_order",aggfunc="size",fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt="d")
plt.xlabel("online_order")            
plt.ylabel("listed_in(type")  
plt.show()            
```
