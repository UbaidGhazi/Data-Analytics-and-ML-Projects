```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("C:\\Users\\HP\\Downloads\\spotify.csv.csv")
df.head(5)
```

```python
df.info()
```

```python
df = df.drop(['artwork_url'], axis = 1)
df = df.drop(['track_url'], axis = 1)
df = df.drop(['track_id'], axis = 1)
```

```python
df['duration_ms'] = df['duration_ms'] / 1000
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
clear_language = df.loc[df['language'] != 'Unknown']
```

```python
cl = clear_language['language'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(cl, labels = cl.index, autopct='%1.1f%%')
plt.title('Language')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='duration_ms', y='popularity')
plt.title('Comparability between song duration and popularity')
plt.xlabel('Time(sec)')
plt.ylabel('Popularity')
plt.show()
```

```python
yearly_popularity =df.groupby('year')['popularity'].sum().reset_index()
yearly_popularity
```

```python
plt.figure(figsize=(10, 6))
plt.bar(yearly_popularity['year'], yearly_popularity['popularity'], color='skyblue')
plt.title('Number of auditions by year')
plt.xlabel('Year')
plt.ylabel('Number of auditions')
plt.xticks(yearly_popularity['year'], rotation = 90)
plt.grid(axis='y')
plt.show()
```

```python
df['popularity'].mean()
```

```python
top_tracks = df[df['popularity'] > 16]
artist_counts = top_tracks.groupby('artist_name')['popularity'].count().reset_index()

artist_counts.columns = ['artist_name', 'track_count']
top_10_artists = artist_counts.sort_values(by='track_count', ascending=False).head(10)
top_10_artists
```

```python
plt.figure(figsize=(10, 12))
plt.pie(top_10_artists['track_count'], labels=top_10_artists['artist_name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 artist by song in top')
plt.axis('equal')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(20, 15))
plt.subplot(2, 2, 2)
sns.heatmap(correlation_matrix, cmap='coolwarm', center=0, annot=True)
plt.title('Correlation Heatmap')
```

```python
correlation_matrix = df[['energy', 'danceability', 'valence','popularity']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Corr between Popularity, Energy, Danceability and Valence')
plt.show()
```

```python
yearly_df = df.groupby('year')[['danceability', 'valence']].mean().reset_index()

plt.figure(figsize=(15, 6))

# Graph by danceability
plt.subplot(1, 2, 1)
plt.plot(yearly_df['year'], yearly_df['danceability'], marker='o', color='blue', label='Danceability')
plt.title('Change in Danceability by year')
plt.xlabel('Year')
plt.ylabel('Danceability')
plt.xticks(yearly_df['year'], rotation=90)
plt.grid()
plt.legend()

#Graph by valence
plt.subplot(1, 2, 2)
plt.plot(yearly_df['year'], yearly_df['valence'], marker='o', color='green', label='Valence')
plt.title('Change Valence by year')
plt.xlabel('Year')
plt.ylabel('Valence')
plt.xticks(yearly_df['year'], rotation=90)
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
most_popular_tracks = df.loc[df.groupby('year')['popularity'].idxmax()]

most_popular_tracks[['year', 'track_name', 'popularity']]
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 2)
plt.plot(most_popular_tracks['year'], most_popular_tracks['popularity'], marker='o', color='red', label='Valence')
plt.title('Change most popularity track by year')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.xticks(most_popular_tracks['year'], rotation=90)
plt.grid()
plt.legend()
```
